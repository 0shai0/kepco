```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from urllib.request import urlretrieve
import time
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://www.google.com/"
driver.get(url)
```

```python
driver.find_elements(By.CSS_SELECTOR,".gb_E")[1].click()
```

```python
elem=driver.find_element(By.NAME,'q')
```

```python
elem.send_keys("고양이")
```

```python
elem.send_keys(Keys.ENTER)
```

```python
elements=driver.find_elements(By.CSS_SELECTOR,".rg_i.Q4LuWd")
for i, elem in enumerate(elements):
    i+=1
    try:
        elem.click()
        driver.implicitly_wait(10) #timesleep처럼 쉬었다가 처리하지만 만일 1초 내에 되면 그전에 처리함
        img_url=driver.find_element(By.CSS_SELECTOR,".sFlh5c.pT0Scc.iPVvYb").get_attribute("src")
        urlretrieve(img_url,"cat/cat"+str(i)+".jpg")
    except:
        print(img_url)
```

```python
script_get_height = "return document.body.scrollHeight;"
current_height = driver.execute_script(script_get_height)
print(current_height)
script_move_height = "window.scrollTo(0,document.body.scrollHeight);"
driver.execute_script(script_move_height)
current_height = driver.execute_script(script_get_height)
print(current_height)
```

```python
elements=driver.find_elements(By.CSS_SELECTOR,".rg_i.Q4LuWd") #클래스는 .으로 처리한다. id는 #(..맞나?)으로 처리한다
```

```python
script_get_height = "return document.body.scrollHeight;"
current_height = driver.execute_script(script_get_height)
while True:
    script_get_height = "return document.body.scrollHeight;"
    driver.execute_script(script_move_height)
    time.sleep(1)
    current_height = driver.execute_script(script_get_height)
    script_move_height = "window.scrollTo(0,document.body.scrollHeight);"
    height = driver.execute_script(script_get_height)
    if current_height == height:
        try:
            driver.find_element(By.CSS_SELECTOR,".LZ4I").click()
        except:
            break
    current_height=height
```
